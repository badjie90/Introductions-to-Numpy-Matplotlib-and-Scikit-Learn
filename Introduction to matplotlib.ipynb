## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x= [10, 20, 30, 40]
y= [100, 200, 300, 400]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. plot data 
ax.plot(x, y)

# 4. customize plot
ax.set(title='simple plot', xlabel='x-axis', ylabel='y-axis')


# 5. save and show (you save the whole figure)
fig.savefig('Introduction to Matplotlib/images/sample-plot.png')



```

```python
x=[1,2,3,4,5,6,7,8,9]
y =[20, 10, 50, 30, 60,30,10,70,50]

fig, ax= plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title='my second plot', xlabel='days', ylabel='weeks')
fig.savefig('Introduction to Matplotlib/images/my second plot.png')
```

## Making a bar plot

```python
# Make a plot from dictionary
nut_butter_prices={'Peanut':25, 'Almond':30, 'Cashew':50, 'coco':20}
fig, ax=plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Bakary\'s Butters', xlabel='Types', ylabel='Prices ($)');
```

```python
# Horizontal Bar
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title= 'Horizontal bar', xlabel='Types', ylabel='Prices ($)');
```

## Two options for subplots

```python
# subplot option1
fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))
# plotting to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(100), np.random.randn(100));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2
fig, ax=plt.subplots(nrows=2, 
                    ncols=2,
                    figsize=(10,5))
# plot to each index
ax[0,0].plot(x, x/2);
ax[0,1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,0].scatter(np.random.randn(100), np.random.randn(100));
ax[1,1].hist(np.random.randn(1000));

```

## which one to use (pyplot vs matplotlib OO method?)
* When plotting something quickly, used the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with pyplot
fig, ax=plt.subplots(figsize=(10, 10))
over_50.plot(kind='scatter', x='age', 
             y='chol',
            c='target', ax=ax);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(5,3))
scatter= ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])
# customize plot
ax.set(title='heart-disease with age over 50', 
      xlabel='Age',
      ylabel='Cholesterol');
# add legend
ax.legend(*scatter.legend_elements(), title='Target');

# add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='dashed');
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
# subplots of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)
scatter=ax0.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target']);

ax0.set(title='Heart Disease Proble',
       ylabel='Cholrol');

ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')


scatter=ax1.scatter(x=over_50['age'],
                   y=over_50['thalach'],
                   c=over_50['target']);

ax1.set(title='Hear Disease', 
       xlabel='age', ylabel='thalach')

ax1.legend(*scatter.legend_elements(), title=('Target'))
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--');
# super title for the whole figure
fig.suptitle('Heart Disease Analysis', fontsize=20, fontweight='bold');
           
```
